```python
from readlif.reader import LifFile
```

```python
filepath = 'data/flim_sirDNA_cellbrite650_zsatck.lif'
```

```python
# Crear una instancia de LifFile y leer el archivo .lif
lif = LifFile(filepath)
```

```python
# Obtener la primera imagen del archivo .lif
image = lif.get_image(0)

# Obtener las dimensiones de la imagen
dimensions = image.dims

# Supongamos que m_idx representa la dimensión de los tiles (ajustar según tu archivo .lif)
m_idx = 2

# Obtener el número de tiles en la dimensión m_idx
if m_idx < len(dimensions):
    n_tiles = dimensions[m_idx]
    print(f"Número de tiles en la dimensión {m_idx}: {n_tiles}")
else:
    print(f"La dimensión {m_idx} está fuera de los límites para esta imagen.")
```
