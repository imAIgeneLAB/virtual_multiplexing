```python
from readlif.reader import LifFile
import numpy as np
from explore_lif import Reader
```

```python
def read_tiled_plane(image_in, channel, plane):
    """Read a plane of a tiled czi/lif file."""
    dstack = []
    if image_in.endswith('.lif'):
    
        m_idx = 3
    
        lif = LifFile(image_in)
        for i in range(0, 200):
            lim = lif.get_image(i)
            available_channels = lim.channels
            print("Available channels:", available_channels)
    
        n_tiles = lim.dims[m_idx]
    
        for m in range(n_tiles):
            data = lim.get_frame(z=0, c=channel, t=0, m=m)
            dstack.append(data)
            
    return dstack
```
