```python
import os
import czifile
import xml.etree.ElementTree as ET
import yaml
```

```python
def guardar_metadatos_czi_en_yaml(archivo_czi):
    try:
        with czifile.CziFile(archivo_czi) as czi:
            metadata = czi.metadata()
            xml_metadata = ET.fromstring(metadata)

            # Convertir XML a diccionario Python
            metadata_dict = xml_to_dict(xml_metadata)

            # Generar nombre de archivo de salida basado en el nombre del archivo CZI
            nombre_base = os.path.splitext(os.path.basename(archivo_czi))[0]
            archivo_salida = f"{nombre_base}_metadatos.yaml"

            # Guardar como YAML
            with open(archivo_salida, 'w', encoding='utf-8') as f:
                yaml.dump(metadata_dict, f, allow_unicode=True, default_flow_style=False)

            print(f"Metadatos guardados exitosamente en formato YAML en: {archivo_salida}")
            return True

    except Exception as e:
        print(f"Error al abrir el archivo CZI o al convertir los metadatos: {e}")
        return False
```

```python
def xml_to_dict(root):
    """
    Función para convertir un ElementTree XML a un diccionario recursivamente.
    """
    d = {}
    for child in root:
        if child:
            if child.tag not in d:
                d[child.tag] = xml_to_dict(child)
            elif type(d[child.tag]) is list:
                d[child.tag].append(xml_to_dict(child))
            else:
                d[child.tag] = [d[child.tag], xml_to_dict(child)]
        elif child.text:
            d[child.tag] = child.text
    return d
```

```python
# Ejemplo de uso
archivo_czi = 'D:/Escritorio/data/data8channels.czi'  # Ruta al archivo CZI
# archivo_salida se definirá dinámicamente según el nombre del archivo CZI

guardado_exitoso = guardar_metadatos_czi_en_yaml(archivo_czi)

if guardado_exitoso:
    print(f"Metadatos del archivo CZI guardados en formato YAML")
else:
    print("Error al intentar guardar los metadatos del archivo CZI en formato YAML.")
```
