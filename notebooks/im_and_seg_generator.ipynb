```python
import cv2
import datetime as dt
import h5py
import numpy as np
import os
from glob import glob 
```

```python
# Definir las dimensiones de las imágenes
HEIGHT = 512
WIDTH = 512
CHANNELS = 3
SHAPE = (HEIGHT, WIDTH, CHANNELS)
```

```python
# Ruta donde se encuentran las imágenes
SOURCE_IMAGES = 'data/docker_vm_test/01_traintest/czi_1000_ac_open_detector_br0%_patches/train'
```

```python
# Obtener lista de imágenes y ordenarlas
images = glob(os.path.join(SOURCE_IMAGES, "*.png"))
images.sort()

NUM_IMAGES = len(images)
```

```python
# Crear la forma para los datasets HDF5
train_shape = (NUM_IMAGES, HEIGHT, WIDTH, CHANNELS)
```

```python
# Especificar una ruta absoluta para guardar los archivos HDF5
output_dir = 'data/prueba/train'
```

```python
# Nombre base para los archivos HDF5
base_name = 'train'
```

```python
print(os.path.join(output_dir, f'{base_name}.im.h5'))
```

```python
# Crear archivo HDF5 para guardar imágenes
hf_images = h5py.File(os.path.join(output_dir, f'{base_name}.im.h5'), 'w')
hf_images.create_dataset(base_name, shape=train_shape, maxshape=train_shape, compression='gzip', compression_opts=9)
```

```python
# Crear archivo HDF5 para guardar segmentaciones
hf_segmentations = h5py.File(os.path.join(output_dir, f'{base_name}.seg.h5'), 'w')
hf_segmentations.create_dataset(base_name, shape=train_shape, maxshape=train_shape, compression='gzip', compression_opts=9)
```

```python
# Iterar sobre las imágenes y guardar en archivos HDF5 separados
for i, img_path in enumerate(images):
    img_name = os.path.basename(img_path)
    img = cv2.imread(img_path)
    img = cv2.resize(img, (WIDTH, HEIGHT), interpolation=cv2.INTER_CUBIC)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Simular una segmentación aleatoria (solo para demostración)
    segmentation = np.random.randint(0, 2, size=(HEIGHT, WIDTH, CHANNELS), dtype=np.uint8)
    
    # Guardar imagen en el archivo HDF5 de imágenes
    hf_images[base_name][i, ...] = img
    
    # Guardar segmentación en el archivo HDF5 de segmentaciones
    hf_segmentations[base_name][i, ...] = segmentation

# Cerrar ambos archivos HDF5 después de guardar todos los datos
hf_images.close()
hf_segmentations.close()
```
