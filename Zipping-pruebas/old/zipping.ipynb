```python
from zipper_owntry import Zipp3r
import numpy as np
import cv2
```

```python
image_in = '/app/data/data3channels-002.tif'
output_path = '/app/Zipping-pruebas/OUTPUT'
```

```python
zipper = Zipp3r(image_in, output_path)

# Create a dictionary containing information on all blocks
zipper.set_block_slices(
    image_shape=[137,6552,6552],
    blocksize=[137,3276,3276],
    margins=[100,100,100]
)
```

```python
# Cut the image according to the block_slices
zipper.split_image()
```

```python
def segmentation_function(image, k=2):
    # Convertir la imagen a un formato que K-means pueda procesar
    Z = image.reshape((-1,3))
    Z = np.float32(Z)

    # Definir los criterios y aplicar K-means
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    _, labels, centers = cv2.kmeans(Z, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    # Convertir de nuevo a la imagen original
    centers = np.uint8(centers)
    segmented_image = centers[labels.flatten()]
    segmented_image = segmented_image.reshape((image.shape))
    
    return segmented_image
```

```python
# Segment the image
zipper.segment_image(
    seg_function=segmentation_function
)
```

```python
# Calculate the zip lines that need to be extracted
zipper.set_zip_blocks()
```

```python
# Resegment these zip lines
zipper.resegment_all_lines()
```

```python
# Relabel all blocks and ziplines to get unique segments
zipper.relabel_blocks_and_lines()

zipper.zip_lines()
```
