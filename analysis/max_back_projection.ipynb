```python
import numpy as np
import imageio
from PIL import Image
```

```python
# Load the TIFF stack
image_path = '/app/data/fake_mixed.tif'
stack = imageio.volread(image_path)

# Sum the images in the stack
summed_image = np.sum(stack, axis=0)

# Normalize the summed image
normalized_image = (summed_image - np.min(summed_image)) / (np.max(summed_image) - np.min(summed_image)) * 255
normalized_image = normalized_image.astype(np.uint8)

# Convert the normalized image to PIL for saving
result_image = Image.fromarray(normalized_image)

# Save the result
result_image.save('/app/prueba/back_projection_146_gray.png')
```
