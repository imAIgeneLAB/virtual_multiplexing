```python
from zipper_owntry import Zipp3r
```

```python
# image_in = '/app/data/data3channels-002.tif'
image_in = '/app/data/data1channel-001.tif'
output_path = '/app/Zipping-pruebas/OUTPUT'
```

```python
zipper = Zipp3r(image_in, output_path)

# Create a dictionary containing information on all blocks
zipper.set_block_slices(
    image_shape=[146, 3788, 3788],
    blocksize=[146, 1894, 1894],
    margins=[0,100,100]
)
```

```python
# Cut the image according to the block_slices
zipper.split_image()
```

```python
import numpy as np
import cv2

def segmentation_function(image):
    # Check if the image is valid
    if image is None:
        raise ValueError("The input image is None. Please provide a valid image.")
    
    # Handle 3D images (multiple slices)
    if len(image.shape) == 3 and image.shape[0] > 1:
        segmented_slices = []
        for i in range(image.shape[0]):
            gray = image[i, :, :]  # Assume each slice is grayscale
            _, segmented = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
            segmented_slices.append(segmented)
        return np.array(segmented_slices)
    
    # Handle 2D grayscale or 3-channel BGR images
    if len(image.shape) == 2:
        gray = image
    elif len(image.shape) == 3 and image.shape[2] == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        raise ValueError(f"Unexpected image format: {image.shape}. Expected a grayscale or BGR image.")
    
    # Apply a fixed threshold
    _, segmented = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    
    return segmented
```

```python
# Segment the image
zipper.segment_image(
    seg_function=segmentation_function
)
```

```python
# Calculate the zip lines that need to be extracted
zipper.set_zip_blocks()
```

```python
# Resegment these zip lines
zipper.resegment_all_lines(margins=[0,100,100])
```

```python
# Relabel all blocks and ziplines to get unique segments
zipper.relabel_blocks_and_lines()

zipper.zip_lines()
```
