```python
import torch
import torchvision
import matplotlib
import numpy
import scipy
import PIL
import skimage as ski
import pyvista #(for visulization)
import h5py

import os
import numpy as np
import h5py
from PIL import Image
```

```python
def images_to_h5(input_folder, output_h5_file):
    # Obtener la lista de archivos PNG en la carpeta
    files = [f for f in os.listdir(input_folder) if f.endswith('.png')]
    files.sort()  # Ordenar los archivos (opcional)

    # Inicializar un array para almacenar todas las imágenes
    images = []

    for filename in files:
        # Leer cada imagen PNG usando PIL
        image_path = os.path.join(input_folder, filename)
        img = Image.open(image_path)
        img = img.convert('RGB')  # Convertir a RGB si es necesario
        img = np.array(img)  # Convertir la imagen a un arreglo NumPy
        images.append(img)

    # Convertir la lista de imágenes a un arreglo NumPy
    images = np.array(images)

    # Crear un archivo HDF5 y almacenar las imágenes dentro
    with h5py.File(output_h5_file, 'w') as hf:
        hf.create_dataset('images', data=images)

    print(f'Archivo HDF5 creado exitosamente: {output_h5_file}')
```

```python
input_folder = 'data/prueba/train'
output_h5_file = 'data/prueba/train/dataset.h5'
```

```python
images_to_h5(input_folder, output_h5_file)
```
